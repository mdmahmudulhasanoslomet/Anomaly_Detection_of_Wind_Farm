```python
# Load the data from the CSV file
df = pd.read_csv("C:/Own use/Thesis/Wind_Turbine/Datasets/Scada_Data(Wind_Turbine)/Wind-Turbine-SCADA-signals-2016/Without_Grd_Column/Latest/Wind-Turbine-SCADA-signals-2016_(T01)_TimeSequence.csv")

```

```python
# Manually select the columns you want to compare
selected_columns = ["Amb_WindDir_Relative_Avg", "Amb_WindDir_Abs_Avg"]  # Change to your desired columns

# Check if selected columns exist in the dataframe
valid_columns = [col for col in selected_columns if col in df.columns]

# Create a single plot for selected histograms
plt.figure(figsize=(10, 6))

# Plot histogram for each selected column
for column in valid_columns:
    sns.histplot(df[column], bins=30, kde=True, label=column, alpha=0.6)

# Titles and labels with larger font size
plt.title(f'Histogram Comparison: {", ".join(valid_columns)}', fontsize=18, fontweight='bold', pad=20)  # Increased font size & padding
plt.xlabel('Value', fontsize=16, fontweight='bold', labelpad=15)  # Larger and bolder labels with extra padding
plt.ylabel('Frequency', fontsize=16, fontweight='bold', labelpad=15)  # Same for y-axis


# Add legend
plt.legend(title="Features", fontsize=15)

# Show plot
plt.show()
```
